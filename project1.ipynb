```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/Iris.csv'


iris_df = pd.read_csv(file_path)


print("Shape of the dataset:", iris_df.shape)
print("Column names:", iris_df.columns)
print("\nData types:\n", iris_df.dtypes)


print("\nSummary Statistics:\n", iris_df.describe())
```

```python

# Histograms
plt.figure(figsize=(10, 6))
sns.histplot(data=iris_df, x='SepalLengthCm', bins=20, kde=True)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris_df.iloc[:, :-1])
plt.title('Boxplot of Iris Features')
plt.ylabel('Measurement (cm)')
plt.xticks(rotation=45)
plt.show()

# Pairplot
sns.pairplot(iris_df, hue='Species')
plt.title('Pairplot of Iris Dataset')
plt.show()
```
